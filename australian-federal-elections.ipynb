```python
# Import dependencies and setup

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as maplotlib
```

```python
# Load and preview dataset – HOR NSW

nsw_df = pd.read_csv('./Resources/HOR_NSW.csv')
nsw_df.head()
```

```python
# Set 'Election' as index
nsw_indexed_df = nsw_df.set_index('Election')
nsw_indexed_df.head()
```

```python
# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Define data
nsw_data = nsw_indexed_df[['ALP', 'LP', 'NAT', 'CP', 'CP/NP', 'UAP', 'GRN', 'DEM', 'DLP', 
                   'ON', 'PROT', 'FT', 'A-S', 'Others']]                
                  
# Plot
nsw_data.plot(figsize=(15,5), kind="bar", stacked=True, alpha=0.9, color=['red', 'blue', 'lime', 'brown',
                                                                         'goldenrod', 'yellow', 'darkgreen', 
                                                                          'aquamarine', 'gold', 'orange', 'navy',
                                                                          'cornflowerblue', 'lightsteelblue', 'purple']
                                                                
             ).legend(loc=(1.04, 0.1), ncol=1, title="Party")

plt.title("Voting distribution, NSW House of Representatives, percentage of vote, 1901 – 2016", fontdict = font_1)
plt.xlabel("Election", fontdict = font_2)
plt.ylabel("% of vote", fontdict = font_2)

# Aesthetics
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

# Show/save
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/australian-federal-elections/Images/nsw.png", 
            transparent=False, dpi=300)
plt.show()
```
